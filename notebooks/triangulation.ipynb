```python
from scipy.spatial import Delaunay as _Delaunay
```

```python
import numpy
numpy.random.seed(123456)
```

```python
points = numpy.random.random((10,2))
```

```python
points
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(points[:,0], points[:,1])
```

```python
dt = _Delaunay(points)
```

```python
dt.simplices
```

```python
edges = []
for simplex in dt.simplices:
    edges.append((simplex[0], simplex[1]))
    edges.append((simplex[1], simplex[2]))
    edges.append((simplex[2], simplex[0]))
```

```python
plt.scatter(points[:,0], points[:,1])
for edge in edges:
    h, t = edge
    x = points[edge,0]
    y = points[edge,1]
    plt.plot(x, y, 'bo', linestyle='--')
plt.show()
```

```python
x
```

```python
edges_set = set([ tuple(sorted(e)) for e in edges])
```

```python
plt.scatter(points[:,0], points[:,1])
for edge in edges_set:
    h, t = edge
    x = points[edge,0]
    y = points[edge,1]
    plt.plot(x, y, 'bo', linestyle='--')
plt.show()
```
